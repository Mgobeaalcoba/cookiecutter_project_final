### process data

```python
processed_df = (
    covid_df
    .select_columns(["Country/Region", "*/*/*"])
    .pivot_longer(
        index="Country/Region",
        names_to="date"
    )
    .transform_column("date", pd.to_datetime)
    .clean_names()
)

processed_df.head() # Converti mi tabla original en una nueva donde pivotee las fechas en una sola columna llamada data
```
